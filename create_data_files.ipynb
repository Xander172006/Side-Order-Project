### Combine csv files

```python
directory = './data/colorchips/'
csv_files = ['power_colorchips.csv', 'range_colorchips.csv', 'support_colorchips.csv', 'lucky_colorchips.csv', 'mobility_colorchips.csv', 'drone_colorchips.csv']

dfs = []

for filename in csv_files:
    file_path = os.path.join(directory, filename)
    df = pd.read_csv(file_path)
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
combined_df.to_csv('./data/colorchips/colorchips.csv', index=False)
print("Combined colorchips CSVs into colorchips.csv")
```
